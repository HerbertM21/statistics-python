# **DISTRIBUCIÓN BINOMIAL**

La distribución binomial es una distribución discreta que se utiliza para modelar el número de éxitos en una secuencia de n ensayos independientes, cada uno de los cuales tiene dos posibles resultados: `éxito o fracaso`. 

La distribución binomial se utiliza para modelar el `número de éxitos` en una secuencia de `n ensayos independientes`, cada uno de los cuales tiene dos posibles resultados: `éxito` o `fracaso`.

>- **Ensayo independiente**: Un ensayo es independiente de los demás si el resultado de un ensayo no afecta el resultado de otro ensayo. Por ejemplo, si lanzamos una moneda 10 veces, cada lanzamiento es independiente de los demás. Si la moneda cae cara en el primer lanzamiento, no afecta la probabilidad de que caiga cara en el segundo lanzamiento.
>
>    Una opinion que no afecte la opinion de otra persona es independiente.

>- **Éxito**: El resultado de un ensayo que se considera favorable. Por ejemplo, si lanzamos una moneda, el resultado es cara, que es un éxito.

>- **Fracaso**: El resultado de un ensayo que no se considera favorable. Por ejemplo, si lanzamos una moneda, el resultado es cruz, que es un fracaso.

>- **Número de éxitos**: El número de ensayos que resultan favorables. Por ejemplo, si lanzamos una moneda 10 veces, el número de éxitos es el número de caras.

>- **Número de fracasos**: El número de ensayos que no resultan favorables. Por ejemplo, si lanzamos una moneda 10 veces, el número de fracasos es el número de cruces.

---

Para calcular la probabilidad de que ocurra un evento, necesitamos de la masa de probabilidad de la distribución binomial. La masa de probabilidad de la distribución binomial se define como:

**$$P(X=k) = \binom{n}{k}p^k(1-p)^{n-k}$$**

Como ya vemos, se usa el coeficiente binomial para calcular la probabilidad de que ocurra un evento. El coeficiente binomial se define como:

$${n\choose k} = \frac{n!}{k!(n-k)!}$$

Donde:

>- **k**: Número de éxitos en n ensayos.
>- **n**: Número de ensayos.
>- **p**: Probabilidad de éxito en cada ensayo.
>- **q**: Probabilidad de fracaso en cada ensayo.

---

**Ejemplo 1**

Supongamos que lanzamos una moneda 10 veces. ¿Cuál es la probabilidad de que salgan 5 caras?

**Solución**

La probabilidad de obtener 5 caras al lanzar una moneda 10 veces se puede calcular utilizando la fórmula de la distribución binomial:

**$$P(X=k) = \binom{10}{5}0.5^5(1-0.5)^{10-5}$$**
 
Donde:

- **k** = 5 (queremos calcular la probabilidad de que haya 5 caras)
- **n** = 10 (lanzamos la moneda 10 veces)
- **p** = 0.5 (la probabilidad de obtener cara en cada lanzamiento es de 0.5)

**$$P(X=5) = \binom{10}{5}0.5^5(1-0.5)^{10-5}$$**

**$$P(X=5) = \frac{10!}{5!(10-5)!}0.5^5(1-0.5)^{10-5}$$**

**$$P(X=5) = \frac{10!}{5!(10-5)!}0.5^5(1-0.5)^{5}$$**

**$$P(X=5) = \frac{10!}{5!(10-5)!}0.5^5(0.5)^{5}$$**

**$$P(X=5) = \frac{10!}{5!(10-5)!}0.5^{10}$$**

**$$P(X=5) = \frac{10!}{5!(10-5)!}0.0009765625$$**

**$$P(X=5) = 0.24609375$$**

---

**Ejemplo 2**

Supongamos que lanzamos una moneda 10 veces. ¿Cuál es la probabilidad de que salgan 5 caras o menos?

**Solución**

La probabilidad de obtener 5 caras o menos al lanzar una moneda 10 veces se puede calcular utilizando la fórmula de la distribución binomial:

**$P(X \leq k) = \sum_{i=0}^{k} \binom{n}{i}p^i(1-p)^{n-i}$**

Donde:

- **k** = 5 (queremos calcular la probabilidad de que haya 5 caras o menos)
- **n** = 10 (lanzamos la moneda 10 veces)
- **p** = 0.5 (la probabilidad de obtener cara en cada lanzamiento es de 0.5)

Para calcular esta suma, podemos usar una calculadora o una tabla de valores de la distribución binomial acumulada. Estos son los resultados para cada valor de i, y su suma acumulada:

$P(X=0) = \binom{10}{0}0.5^0(1-0.5)^{10-0} = 0.0009765625$

$P(X=1) = \binom{10}{1}0.5^1(1-0.5)^{10-1} = 0.009765625$

$P(X=2) = \binom{10}{2}0.5^2(1-0.5)^{10-2} = 0.0439453125$

$P(X=3) = \binom{10}{3}0.5^3(1-0.5)^{10-3} = 0.1171875$

$P(X=4) = \binom{10}{4}0.5^4(1-0.5)^{10-4} = 0.205078125$

$P(X=5) = \binom{10}{5}0.5^5(1-0.5)^{10-5} = 0.24609375$

Por lo tanto, la suma acumulada es:

$P(X \leq 5) = 0.0009765625 + 0.009765625 + 0.0439453125 + 0.1171875 + 0.205078125 + 0.24609375 = 0.623046875$

**Ejemplo 3**

Supongamos que lanzamos una moneda 10 veces. ¿Cuál es la probabilidad de que salgan 5 caras o más?

**Solución**

La probabilidad de obtener 5 caras o más al lanzar una moneda 10 veces se puede calcular utilizando la fórmula de la distribución binomial:

**$P(X \geq k) = \sum_{i=k}^{n} \binom{n}{i}p^i(1-p)^{n-i}$**

Donde:

- **k** = 5 (queremos calcular la probabilidad de que haya 5 caras o más)
- **n** = 10 (lanzamos la moneda 10 veces)
- **p** = 0.5 (la probabilidad de obtener cara en cada lanzamiento es de 0.5)

Para calcular la probabilidad de obtener menos de 5 caras, podemos usar la fórmula de la distribución binomial:

$P(X \geq 5) = \sum_{i=5}^{10} \binom{10}{i}0.5^i(1-0.5)^{10-i}$

$P(X \geq 5) = \binom{10}{5}0.5^5(1-0.5)^{10-5} + \binom{10}{6}0.5^6(1-0.5)^{10-6} + \binom{10}{7}0.5^7(1-0.5)^{10-7} + \binom{10}{8}0.5^8(1-0.5)^{10-8} + \binom{10}{9}0.5^9(1-0.5)^{10-9} + \binom{10}{10}0.5^{10}(1-0.5)^{10-10}$

$P(X \geq 5) = 0.24609375 + 0.205078125 + 0.1171875 + 0.0439453125 + 0.009765625 + 0.0009765625$

$P(X \geq 5) = 0.623046875$

## **DEMOSTRACIÓN PROBABILIDAD BINOMIAL**

Para demostrar que la ``probabilidad binomial es de 0.5``, utilizamos la ``ley de los grandes números``. Esta ley dice que la ``probabilidad`` de que la ``proporción`` de ``eventos favorables`` en un número grande de ensayos sea aproximadamente igual a la probabilidad de que ocurra un evento favorable.

En nuestro ejemplo, la probabilidad de que salga cara en un lanzamiento de moneda es de 0.5. Si lanzamos la moneda 100 veces, la probabilidad de que salgan 50 caras es de 0.5. Si lanzamos la moneda 1000 veces, la probabilidad de que salgan 500 caras es de 0.5. Si lanzamos la moneda 10000 veces, la probabilidad de que salgan 5000 caras es de 0.5. Y así sucesivamente.

**EJEMPLO EN PYTHON**




```python
import numpy as np
import matplotlib.pyplot as plt

n_experiments = 1200
results = []

caras = 0
for experiment in range(n_experiments):
    monedas = np.random.randint(0, 2, size=2)
    caras += np.sum(monedas == 1)
    longitud = len(monedas)
    # Formula proporción = caras / (numeros de lanzamientos * numero de experimento actual) 
    # Formula proporción = caras / (longitud de la muestra * numero de experimento actual + 1) (empieza desde 0, por eso) 
    proporción = caras / (longitud * (experiment + 1))
    results.append(proporción)

# LEY DE LOS GRANDES NÚMEROS
mean_results = np.mean(results)
print('Ley de los grandes números: ', mean_results)

# Graficar los resultados
# Graficar los resultados de todos los experimentos juntos
# plt.plot(10, [lista de proporciones de cara], '-o')
plt.plot(range(1, n_experiments+1), results, '-o')
plt.title('Proporción de caras en lanzamiento de moneda')
plt.axhline(mean_results, color='r', linestyle='--', label='Ley de los grandes números')
plt.xlabel('Experimentos')
plt.ylabel('Proporción de caras')
plt.ylim([0, 1])
plt.xlim([0, n_experiments+1])
plt.legend()
plt.show()

# La convergencia de 0.5 en una distribucion binomial se refiere a la probabilidad de que salga cara en una moneda, 
# es decir, 0.5. En este caso, la ley de los grandes numeros nos dice que la proporción de caras en cada experimento
# se acerca a 0.5 conforme aumenta el numero de experimentos.

# ES IMPORTANTE QUE LA LEY DE GRANDES NUMEROS NO SE DESCARTA QUE ALGUNOS RESULTADOS SEAN MUY DISTINTO A SU CONVERGENCIA
# SIEMPRE LOS PRIMEROS EXPERIMENTOS SON LOS QUE MAS SE DESVIAN DE LA CONVERGENCIA
# LOS QUE ESTAN MAS CERCA DE LA CONVERGENCIA SON LOS ULTIMOS EXPERIMENTOS
```

## **DISTRIBUCIÓN DE BERNOULLI**

La distribución de Bernoulli es una distribución discreta parecida a la distribución binomial, pero con un solo ensayo.

La distribución de Bernoulli se utiliza para modelar el resultado de un experimento aleatorio con dos posibles resultados, como por ejemplo:

- **Lanzar una moneda y obtener cara o cruz.**
- **Tirar un dado y obtener un número par o impar.**
- **Sacar una carta de una baraja y obtener un as o no.**

Es casi lo mismo que hicimos en el ejemplo 2 y 3, calculando la probabilidad de cada ensayo, con la diferencia de que no hay una sumatoria de ensayos, **sino que solo hay un ensayo.**

**La formula es la siguiente:**

**$P(X=k) = p^k(1-p)^{1-k}$**

Donde:

- **k** = Resultado del ensayo (0 o 1)
- **p** = Probabilidad de obtener el resultado k (0.5 en el ejemplo de la moneda)
- **X** = Variable aleatoria que representa el resultado del ensayo
- **P(X=k)** = Probabilidad de obtener el resultado k para la variable aleatoria X
