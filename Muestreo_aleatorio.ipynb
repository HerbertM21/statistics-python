# **Técnica de muestreo aleatorio y la extrapolación de la muestra para estimar la población completa.**

**Problema de Gauss:** Una sucesión de 100 números aleatorios, estimar la población completa. Sin utilizar la solución de Gauss.

**La solución propuesta es:** 
- **Paso 1**: Se utiliza una muestra aleatoria del 20% de la población, 
- **Paso 2**: Se calcula la suma de los números en esta muestra 
- **Paso 3**: Se multiplica por 5 para obtener una estimación de la suma total de la población completa. 

Esta estimación se basa en la suposición de que la **muestra aleatoria** es **representativa** de la población completa, y por lo tanto, la suma de los números en la muestra es un buen estimador de la suma total de los números en la población completa.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el tamaño de la población
tamano_poblacion = 100

# Crear una muestra aleatoria de tamaño 20
tamano_muestra = int(tamano_poblacion * 0.2) # 20% de la población

# Crea una lista de 20 numeros aleatorios entre 1 y 100
muestra_20 = np.random.choice(range(1, tamano_poblacion+1), tamano_muestra)

# Calcular la suma de la muestra
suma_muestra = np.sum(muestra_20)

# Extrapolación de la muestra para estimar la suma total
suma_estimada = suma_muestra * 5 # (tamaño_poblacion / tamaño_muestra)

# Imprimir los resultados
print("Suma de la muestra:", suma_muestra)
print("Estimación de la suma total:", suma_estimada)

# Crear un histograma de la muestra
plt.hist(muestra_20, bins=range(1, tamano_muestra+2), alpha=0.5)
plt.axvline(x=suma_muestra, color='g', linestyle='--', label='Suma de la muestra')
plt.axvline(x=suma_estimada, color='b', linestyle='--', label='Estimación de la suma total')
plt.legend()
plt.show()
```

## **¿Donde se utiliza el muestreo aleatorio?**

El muestreo aleatorio se utiliza en una amplia variedad de campos para obtener **muestras representativas de una población**. Algunos ejemplos de casos en los que se utiliza el muestreo aleatorio son:

- **Encuestas de opinión**: en las que se busca conocer la opinión de una población sobre un tema determinado.
- **Investigación de mercado**: en la que se busca conocer las preferencias de los consumidores sobre un producto o servicio.
- **Investigación científica**: en la que se busca obtener una muestra representativa de la población de estudio para realizar experimentos o estudios estadísticos.
- **Estudios de calidad**: en los que se busca verificar la calidad de un producto o servicio a través de una muestra representativa de la población de producción.
- **Auditorías financieras**: en las que se busca verificar la integridad financiera de una empresa a través de una muestra representativa de sus transacciones financieras.

En general, el muestreo aleatorio se utiliza en cualquier situación en la que se necesite obtener una muestra representativa de una población y en la que sea necesario **evitar sesgos o influencias externas en la selección de la muestra.**

## **Muestreo aleatorio simple**

El **muestreo aleatorio simple** es un **método de selección** que se utiliza para elegir muestras de una población. En este método, todas las **unidades de análisis (UAs)** de la población tienen la misma probabilidad de ser elegidas.

Una muestra aleatoria simple de **tamaño n** de una **variable aleatoria X con función de distribución FX**, es un conjunto de **n variables aleatorias [X1, X2, X3, ..., Xn]** que son independientes y tienen la **misma distribución que X** de acuerdo a la **función de distribución FX**. 

En otras palabras, cada **variable aleatoria Xi** en la muestra se selecciona de manera independiente y con la misma probabilidad que las demás, lo que asegura que la muestra sea representativa de la población y que no haya sesgos en la selección de la muestra. Entonces por eso se elige aleatoriamente los representantes en la muestra.
