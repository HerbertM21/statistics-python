## **ESTIMADOR POR INTERVALOS**

El cálculo de un **estimador por intervalos** implica la construcción de un ``intervalo de confianza`` alrededor de un **estimador puntual**. 


El estimador por intervalos es una forma de **determinar el rango de valores** donde es más probable o confiable que se encuentre el parámetro de la población, a partir de la información de una muestra. El cálculo dependerá de:

* El **tamaño de la muestra**.
* El **nivel de confianza**.
* La **distribución de la población**.

---

- El **nivel de confianza** es la probabilidad de que el intervalo de confianza contenga el parámetro de la población. Por ejemplo, si el nivel de confianza es del 95%, entonces el intervalo de confianza contendrá el parámetro de la población con una probabilidad del 95%.

---

- El **tamaño de la muestra** es el número de observaciones en la muestra. El tamaño de la muestra afecta el ancho del intervalo de confianza. Un intervalo de confianza más ancho tendrá un nivel de confianza más bajo. Un intervalo de confianza más estrecho tendrá un nivel de confianza más alto.

---

- La **distribución de la población** es la distribución de la variable aleatoria de la población. El tipo de distribución de la población afecta el ancho del intervalo de confianza. Un intervalo de confianza más ancho tendrá un nivel de confianza más bajo. Un intervalo de confianza más estrecho tendrá un nivel de confianza más alto.

---

- El **ancho del intervalo de confianza** es la diferencia entre los ``límites superior`` e ``inferior`` del ´´intervalo de confianza´´. El ancho del intervalo de confianza es una medida de la precisión del intervalo de confianza. 

    Un intervalo de confianza más ancho tendrá un nivel de confianza más bajo. Un intervalo de confianza más estrecho tendrá un nivel de confianza más alto.

---

### **CÁLCULO DEL ESTIMADOR POR INTERVALOS - DIST. NORMAL**

El cálculo del estimador por intervalos para una distribución normal se realiza a partir de la siguiente fórmula:

$$\bar{x} \pm z_{\alpha/2} \frac{\sigma}{\sqrt{n}}$$

donde:

* $\bar{x}$ es la media muestral.
* $\sigma$ es la desviación estándar de la población.
* $n$ es el tamaño de la muestra.
* $z_{\alpha/2}$ es el valor crítico de la distribución normal estándar.

**PASOS PARA CALCULAR**:

**1.** Obtén una muestra aleatoria de la población y calcula su media muestral, denotada como x̄.

**2.** Determina el nivel de confianza deseado para el intervalo de confianza, generalmente se utiliza un nivel de confianza del 95% o del 99%.

**3.** Encuentra el valor crítico de la distribución normal estándar que corresponde al nivel de confianza seleccionado. Por ejemplo, si el nivel de confianza es del 95%, el valor crítico es 1.96. Si el nivel de confianza es del 99%, el valor crítico es 2.58.

**4.** Calcula el error estándar de la media utilizando la fórmula:

    SE = σ / sqrt(n)

    Donde σ es la desviación estándar poblacional conocida (si no se conoce, se puede utilizar la desviación estándar muestral s como una estimación), n es el tamaño de la muestra.

**5.** Construye el intervalo de confianza utilizando la fórmula:

    x̄ ± z * SE

    Donde x̄ es la media muestral, z es el valor crítico de la distribución normal estándar correspondiente al nivel de confianza seleccionado, y SE es el error estándar de la media calculado en el paso anterior.

**6.** Interpreta el intervalo de confianza. Por ejemplo, si el intervalo de confianza es (10, 20), se puede interpretar como "con un nivel de confianza del 95%, la media poblacional se encuentra entre 10 y 20".
