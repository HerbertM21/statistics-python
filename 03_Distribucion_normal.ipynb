# **DISTRIBUCIONES DE PROBABILIDADES**

Son **funciones matematicas** que nos permiten **modelar** el comportamiento de una **variable aleatoria**. Describen la probabilidad de que ocurra cada valor en una población o muestra.


> - **VARIABLE ALEATORIA**: Es una variable que puede tomar cualquier valor dentro de un conjunto de valores posibles. Por ejemplo, el resultado de lanzar un dado es una variable aleatoria, ya que puede tomar cualquier valor entre 1 y 6.

> - **POBLACIÓN**: Es el conjunto de todos los posibles valores que puede tomar una variable aleatoria.

> - **MUESTRA**: Es un subconjunto de la población.

Hay dos tipos de distribuciones de probabilidad:

- **Distribuciones continuas**: La `variable aleatoria` puede tomar cualquier valor dentro de un intervalo/rango. Por ejemplo, la altura de una persona.

- **Distribuciones discretas**: La `variable aleatoria` puede tomar solo valores específicos. Por ejemplo, el resultado de lanzar un dado.

> - **PASOS PARA CONSTRUIR UNA DISTRIBUCION DE PROBABILIDADES**:
>
> ``1. Identificar la variable aleatoria X que se va a estudiar``
>
> ``2. Identificar los posibles valores que puede tomar X.``
>
> ``3. Asignar una probabilidad a cada valor de X.``
>
> ``4. Verificar que la suma de las probabilidades sea igual a 1.``

---

### **1. DISTRIBUCIÓN NORMAL**:

Es de tipo continua y simétrica. Se usa para modelar variables aleatorias que siguen una distribución normal.

La distribución normal tiene dos ``parametros``:

- **Media**: Es el valor central de la distribución. Se representa con la letra griega mu (μ).
- **Desviación estándar**: Es la medida de dispersión de los datos. Se representa con la letra griega sigma (σ).



En un plano cartesiano, la distribución normal se representa con una curva que tiene forma de campana. La media es el punto más alto de la curva y la desviación estándar es la distancia entre la media y los puntos de inflexión.

> - **EJE X**: Representa los valores de la variable aleatoria.
> - **EJE Y**: Representa la probabilidad de que la variable aleatoria tome ese valor.

> - **PROPIEDADES DE LA DISTRIBUCIÓN NORMAL**:
> - La distribución normal es simétrica.
> - La media, la mediana y la moda son iguales.
> - El 68% de los datos se encuentran a una desviación estándar de la media.
> - El 95% de los datos se encuentran a dos desviaciones estándar de la media.
> - El 99.7% de los datos se encuentran a tres desviaciones estándar de la media.


---

##### **EJEMPLO 1 CON PYTHON**:

Como la distribución normal es continua, podemos usar el siguiente ejemplo:

**¿Cuál es la probabilidad de que la altura de una persona sea de 1.80 metros?**

Los datos son:
- **Poblacion**: 1000 personas.
- **Variable aleatoria**: 1.80 metros.
- **Posibles valores**: [170, 181]
- **Asignar una probabilidad**: Funcion de densidad de probabilidad (f(x)).
- **Verificar que la suma de las probabilidades sea igual a 1**: 1.

**Problemas:**

- No tenemos los datos de la población. Por lo cual, necesitamos calcular el ``estimador puntual``: **la media y la desviación estándar.**

**¿Como se calcula el estimador puntual?**

Mediante una muestra de la población. Por ejemplo, si tenemos 100 personas, podemos calcular la media y la desviación estándar de la muestra.

Una mejor manera es utilizar la verosimilitud, pero se verá en otro **notebook**.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate

# Instuimos que distribución queremos para la variable aleatoria
# Distribución normal: Densidad de probabilidad de una variable aleatoria

def densidad_normal(x, mu, sigma):
    return 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5*((x-mu)/sigma)**2)

# Creamos la muestra aleatoria de 100 personas

x = np.linspace(0, 250, 100)

# Estimamos la media y la desviación típica de la muestra mediante el metodo de media muestral y desviación típica muestral
# En otro caso, se utilizaria la verosimilitud maxima para estimar la media y la desviación típica
mu = np.mean(x)
sigma = np.std(x)
VA = 180

# Calculamos la densidad de probabilidad de la variable aleatoria

y = densidad_normal(x, mu, sigma)

# Calculamos la integral de la densidad de probabilidad de la variable aleatoria
# integrate.quad sirve para calcular la integral de una función, quad es un método de integración numérica de cuadratura adaptativa de Gauss-Kronrod

a = 179
b = 181
# se usa args para pasar los parámetros de la función densidad_normal
# dado que se necesita de mu y sigma para calcular la funcion de la densidad de probabilidad
probabilidad, error = integrate.quad(densidad_normal, a, b, args=(mu, sigma))

# Graficamos la densidad de probabilidad de la variable aleatoria

plt.plot(x, y)
plt.title('Densidad de probabilidad de una variable aleatoria')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad de probabilidad')
plt.show()

# Graficamos la probabilidad de la variable aleatoria

plt.plot(x, y)
plt.fill_between(x, y, where=(x>=a) & (x<=b), color='red')
plt.title('Probabilidad de una variable aleatoria')
plt.xlabel('Altura (cm)')
plt.ylabel('Probabilidad')
plt.show()

print('La probabilidad de que una persona mida entre 179 y 181 cm es de: ', probabilidad*100,'%')
```
