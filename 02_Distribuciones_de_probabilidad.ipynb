# **DISTRIBUCIONES DE PROBABILIDADES**

Son **funciones matematicas** que nos permiten **modelar** el comportamiento de una **variable aleatoria**. Describen la probabilidad de que ocurra cada valor en una población o muestra.

Hay dos tipos de distribuciones de probabilidad:

- **Distribuciones continuas**: La `variable aleatoria` puede tomar cualquier valor dentro de un intervalo/rango. Por ejemplo, la altura de una persona.

- **Distribuciones discretas**: La `variable aleatoria` puede tomar solo valores específicos. Por ejemplo, el resultado de lanzar un dado.

---

> **DISTRIBUCIONES EXISTENTES**
>
>> **- Distribución normal (Gaussiana):** es una distribución continua que se utiliza para modelar muchos fenómenos naturales, como la altura de las personas, el peso, el tiempo de reacción, etc. Tiene una forma de campana y se caracteriza por su media y su desviación estándar.
>
>> **- Distribución binomial (Bernoulli):** es una distribución discreta que se utiliza para modelar eventos que solo tienen dos resultados posibles, como éxito o fracaso, cara o cruz, etc. Se caracteriza por su número de ensayos y su probabilidad de éxito.
>
>> **- Distribución de Poisson:** es una distribución discreta que se utiliza para modelar eventos raros o poco frecuentes, como accidentes de tráfico, errores en un informe, etc. Se caracteriza por su tasa media de ocurrencia.
>
>> **- Distribución uniforme:** es una distribución continua que se utiliza cuando todos los valores posibles tienen la misma probabilidad de ocurrir, como en el lanzamiento de un dado o una moneda. Se caracteriza por su valor mínimo y máximo.
>
>> **- Distribución exponencial:** es una distribución continua que se utiliza para modelar el tiempo entre eventos, como el tiempo entre llegadas de clientes a un negocio. Se caracteriza por su tasa media de ocurrencia.
>
>> **- Distribución por función hypergeometrica:** es una distribución discreta que se utiliza para modelar eventos donde se extraen elementos de un conjunto sin reemplazo, como en el caso de extraer bolas de una urna o cartas de una baraja. Se caracteriza por el número total de elementos, el número de elementos en la categoría de interés y el tamaño de la muestra extraída.

---

## **DENSIDAD DE PROBABILIDAD - DISTRIBUCION DE MASA**

La **densidad de probabilidad** es una función que nos permite saber la forma de la distribución de probabilidad de una variable aleatoria. Esto nos ayuda a graficar la distribución de probabilidad.

Una de las caracteristicas de esto, es el `calculo de los estimadores puntuales` de la formula de densidad de probabilidad. Esto se hace eficientemente mediante la `maximización de la verosimilitud`.

> `La densidad de probabilidad` = `Forma de la distribución de probabilidad.`

La **densidad de probabilidad** de una variable aleatoria `X` se define como:

$$f(x) = \frac{dP(X=x)}{dx}$$

- Siendo `DP` la **densidad de probabilidad** de la variable aleatoria `X` (Una función matemática).
- Siendo `x` el valor de la variable aleatoria `X` (Una lista de valores).
- Siendo `P(X=x)` es la probabilidad de que la variable aleatoria `X` tome el valor `x`.
- Siendo `dx` es el tamaño del intervalo en el que se calcula la probabilidad.

Esto sirve solamente para distribuciones continuas. Para distribuciones discretas, la probabilidad de que la variable aleatoria tome un valor `x` es igual a `0`.

Las formulas de densidad de probabilidad son diferentes para cada distribución de probabilidad:

>
>>- **Distribución normal (Gaussiana):** $f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}$
>>
>> Siendo $\mu$ la media y $\sigma$ la desviación estándar.
>
>>- **Distribución uniforme:** $f(x) = \frac{1}{b-a}$
>>
>> Siendo `a` el valor mínimo y `b` el valor máximo.
>
>>- **Distribución exponencial:** $f(x) = \lambda e^{-\lambda x}$
>>
>> Siendo $\lambda$ la tasa media de ocurrencia.
>

## **MASA DE PROBABILIDAD**

La **masa de probabilidad** es una función que nos permite calcular la probabilidad de que una variable aleatoria tome un valor `especifico`.

La **masa de probabilidad** de una variable aleatoria `X` se define como:

$$f(x) = MP(X=x)$$

- Siendo `MP` la **masa de probabilidad** de la variable aleatoria `X` (Una función matemática).
- Siendo `x` el valor de la variable aleatoria `X` (Una lista de valores).
- Siendo `P(X=x)` es la probabilidad de que la variable aleatoria `X` tome el valor `x`.

Esto sirve solamente para distribuciones discretas. Para distribuciones continuas, la probabilidad de que la variable aleatoria tome un valor `x` es igual a `0`.

Las formulas de masa de probabilidad son diferentes para cada distribución de probabilidad:

>>- **Distribución binomial (Bernoulli):** $f(x) = \binom{n}{x} p^x (1-p)^{n-x}$
>>
>> Siendo `n` el número de ensayos y `p` la probabilidad de éxito.
>
>>- **Distribución de Poisson:** $f(x) = \frac{\lambda^x e^{-\lambda}}{x!}$
>>
>> Siendo $\lambda$ la tasa media de ocurrencia.
>
>>- **Distribución por función hypergeometrica:** $f(x) = \frac{\binom{N}{x} \binom{N_0}{n-x}}{\binom{N+N_0}{n}}$
>>
>> Siendo `N` el número total de elementos, `N0` el número de elementos en la categoría de interés y `n` el tamaño de la muestra extraída.

## **INTEGRAL DE LA DENSIDAD DE PROBABILIDAD**

La **integral de la densidad de probabilidad** es una función que nos permite calcular la probabilidad de que una variable aleatoria tome un valor en un `intervalo dado`.

La **integral de la densidad de probabilidad** de una variable aleatoria `X` se define como:

$$F(x) = \int_{-\infty}^{x} f(x) dx$$

- Siendo `F` la **integral de la densidad de probabilidad** de la variable aleatoria `X` (Una función matemática).
- Siendo `x` el valor de la variable aleatoria `X` (Una lista de valores).
- Siendo `f(x)` es la **densidad de probabilidad** de la variable aleatoria `X`.
- Siendo `dx` es el tamaño del intervalo en el que se calcula la probabilidad.



**Por ejemplo, en el caso de la distribución normal, la integral de la densidad de probabilidad es:**

$$P(a \leq X \leq b) = \int_{a}^{b} \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}} dx$$

Si queremos saber si un valor `1.80` cae dentro de la distribución normal, podemos calcular la integral de la densidad de probabilidad para ese valor:	

$$P(a \leq 1.81 \leq b) = \int_{1.79}^{1.81} \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(1.81-\mu)^2}{2\sigma^2}} dx$$

Entonces como sabemos cual es sigma y mu, podemos elegir el método mas eficiente para calcular estimadores puntuales desconocidos, el cual es el método de la `verosimilitud maxima`.

---

### **APUNTES**

- Para indicar la probabilidad de que un elemento especifico sea elegido, se utiliza la **masa de probabilidad**.
- **No podemos indicar la probabilidad de que un elemento especifico sea elegido** en una distribución continua, por lo tanto la densidad de probabilidad ni la masa de probabilidad no se pueden utilizar para estos casos.
- La **densidad de probabilidad** permite graficar la forma de la distribución de probabilidad.
- Necesitamos la integral para calcular la probabilidad de que una variable aleatoria tome un valor en un intervalo dado en la densidad de probabilidad.
- Debemos utilizar la integral de la **densidad de probabilidad** para calcular la probabilidad de que un elemento caiga en un intervalo de valores.
- Se debe `instuir` que función de densidad o de masa de probabilidad se utiliza para cada conjunto de datos que se analiza.
- Se utiliza la **verosimilitud maxima** para encontrar los parametros de la densidad de probabilidad que se intuya.
